```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
nike = yf.Ticker("NKE")
# get historical market data
nike_hist =nike.history(period="max")
nike_hist
```

```python
nike = yf.Ticker("NKE")
# get historical market data
nike_hist =nike.history(period="max")
nike_hist = nike_hist[["Close"]]
nike_hist = nike_hist.rename(columns={"Close": "NIKE Close"})
nike_hist
```

```python
lenovo = yf.Ticker("LNVGY")
# get historical market data
lenovo_hist =lenovo.history(period="max")
lenovo_hist = lenovo_hist[["Close"]]
lenovo_hist = lenovo_hist.rename(columns={"Close": "LENOVO Close"})
lenovo_hist
```

```python
#merge
data = pd.concat([nike_hist, lenovo_hist], axis=1)
data = data["2013-01-01":"2023-12-31"]
data
```

```python
data.to_csv("data.csv")
```
